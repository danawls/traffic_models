```python
# 나머지는 KNN(범주형)을 통해 보정
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# 범주형 데이터를 숫자로 변환
cam_encoded = cam.apply(lambda x: pd.factorize(x)[0])

# KNN Imputer 생성 (K=2로 설정)
imputer = KNNImputer(n_neighbors=2)

# 결측치 대체
cam_imputed = pd.DataFrame(imputer.fit_transform(cam_encoded), columns=cam.columns)

# 숫자로 변환된 데이터를 다시 문자로 변환
for column in cam.columns:
    cam_imputed[column] = pd.Categorical.from_codes(cam_imputed[column].astype(int), categories=pd.factorize(cam[column])[1])

print(cam_imputed)
```

# 보정 실폐

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# 범주형 데이터를 원-핫 인코딩으로 변환
cam_encoded = pd.get_dummies(cam)

# KNN Imputer 생성 (K=2로 설정)
imputer = KNNImputer(n_neighbors=2)

# 결측치 대체
cam_imputed = pd.DataFrame(imputer.fit_transform(cam_encoded), columns=cam_encoded.columns)

# 원-핫 인코딩된 데이터를 다시 범주형 데이터로 변환
for column in cam.columns:
    # 각 범주별로 가장 높은 값을 갖는 열의 이름으로 원래 값 복원
    if pd.api.types.is_categorical_dtype(cam[column]):
        cam_imputed[column] = cam_imputed.filter(like=column).idxmax(axis=1).apply(lambda x: x.split('_')[1])

print(cam_imputed)
```
