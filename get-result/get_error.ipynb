```python
def get_value(a):
    ab = []
    result = {'mse':[], 'rmse':[], 'mae':[], 'r2':[], 'mape':[]}
    for i in a:
       ab.append(i.split(','))
    for k in ab:
        for i in k:
            h = i.split(':')[1][1:]
            ab[ab.index(k)][k.index(i)] = h
    for k in ab:
        for i in range(len(k)):
            if i == 0:
                result['mse'].append(k[i])
            if i == 1:
                result['rmse'].append(k[i])
            if i == 2:
                result['mae'].append(k[i])
            if i == 3:
                result['r2'].append(k[i])
            if i == 4:
                result['mape'].append(k[i])
                
    for i in result:
        for k in result[i]:
            result[i][result[i].index(k)] = float(k)
    
    return result
```

```python
def get_mean(a):
    result = {}
    for i in a:
        if i == 'mse':
            result['mse'] = np.mean(a[i])
        if i == 'rmse':
            result['rmse'] = np.mean(a[i])
        if i == 'mae':
            result['mae'] = np.mean(a[i])
        if i == 'r2':
            result['r2'] = np.mean(a[i])
        if i == 'mape':
            result['mape'] = np.mean(a[i])
            
    for i in result:
        result[i] = np.round(result[i], 4)
    
    return result
```

```python
def activate(a):
    ab = get_value(a)
    ba = get_mean(ab)
    return ba
```

```python
def get_value2(a):
    ab = []
    result = {'mse': [], 'rmse': [], 'mae': [], 'r2': [], 'mape': []}
    for i in a:
        ab.append(i.split(','))
    for k in ab:
        for i in k:
            h = i.split(':')[1][1:]
            ab[ab.index(k)][k.index(i)] = h
    for k in ab:
        for i in range(len(k)):
            if i == 0:
                result['mae'].append(k[i])
            if i == 1:
                result['mse'].append(k[i])
            if i == 2:
                result['rmse'].append(k[i])
            if i == 3:
                result['r2'].append(k[i])
            if i == 4:
                result['mape'].append(k[i])

    for i in result:
        for k in result[i]:
            result[i][result[i].index(k)] = float(k)

    return result


def get_mean2(a):
    result = {}
    for i in a:
        if i == 'mse':
            result['mse'] = np.mean(a[i])
        if i == 'mae':
            result['mae'] = np.mean(a[i])
        if i == 'rmse':
            result['rmse'] = np.mean(a[i])
        if i == 'r2':
            result['r2'] = np.mean(a[i])
        if i == 'mape':
            result['mape'] = np.mean(a[i])
            
    for i in result:
        result[i] = np.round(result[i], 4)

    return result


def activate2(a):
    ab = get_value2(a)
    ba = get_mean2(ab)
    return ba
```
